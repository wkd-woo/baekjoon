```python
n = int(input())

dp = [0] * (1000000 + 1)
memo = [[i] for i in range(1000000 + 1)]
for i in range(2, 1000000 + 1):
    dp[i] = dp[i - 1] + 1

    if i % 2 == 0:
        dp[i] = min(dp[i], dp[i // 2] + 1)
        if dp[i // 2] + 1 > dp[i]:
            memo[i].append(i - 1)
        else:
            memo[i].append(i // 2)
    if i % 3 == 0:
        dp[i] = min(dp[i], dp[i // 3] + 1)
        if dp[i // 3] + 1 > dp[i]:
            memo[i].append(i - 1)
        else:
            memo[i].append(i // 3)

result = [n]
while True:
    if n == 1:
        break

    print(*memo[n], end=' ')
    n = dp[n]
```
