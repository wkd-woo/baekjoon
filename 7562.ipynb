```python
from collections import deque

#t = int(input())

# 체스 좌표평면
dx = [1, -1, 1, -1, 2, 2, -2, -2]
dy = [2, 2, -2, -2, 1, -1, 1, -1]

l = int(input())
graph = [[0] * l for _ in range(l)]
visited = [[0] * l for _ in range(l)]

start = tuple(map(int, input().split()))
end = tuple(map(int, input().split()))

graph[start[0]][start[1]] = 1
graph[end[0]][end[1]] = 1

queue = deque()
queue.append(start)

answer = 0
breaker = False
answer = int(1e9)
while queue:

    if breaker:
        break

    cnt = 0

    x, y = queue.popleft()
    if (x, y) == end:
        print(cnt)
        break

    for i in range(8):
        nx = x + dx[i]  # 상하좌우 돌면서 확인
        ny = y + dy[i]
        # 그래프 범위에서 벗어나면 생략함
        if nx < 0 or nx >= l or ny < 0 or ny >= l:
            continue

        # 현 위치에서 그래프 값이 0이면
        if graph[nx][ny] == 0 and not visited[nx][ny]:
            queue.append((nx, ny))  # 현 위치 좌표를 queue에 추가
            #graph[nx][ny] = -1
            visited[nx][ny] = 1
            cnt += 1
        elif graph[nx][ny] == 1 and nx == end[0] and ny == end[1]:
            queue.clear()
            breaker = True
            answer = min(answer, cnt)
            break

print(answer)
```
